## Detecting Malicious URL With Machine Learning In Julia
#### Requirement
+ Pkg.add("DataFrames")
+ Pkg.add("ScikitLearn")

```julia
using DataFrames
using ScikitLearn: fit!,fit_transform!,predict,@sk_import
```

```julia
# Loading Data of Urls
urls = readtable("urldata.csv")
```

```julia
head(urls,10)
```

```julia
size(urls)
```

```julia
# X Features and Y Labels
X = convert(Array,urls[:url])
```

```julia
y = convert(Array,urls[:label])
```

```julia
url_list = []
for i in urls[:url]
    push!(url_list,i)
end
```

```julia
println(url_list)
```

```julia
# Vectorization 
@sk_import feature_extraction.text: TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```julia
Xf = fit_transform!(vectorizer,url_list)
```

### Split Training and Testting Data


```julia
@sk_import model_selection: train_test_split
```

```julia
X_train, X_test, y_train, y_test = train_test_split(Xf, y, test_size=0.2, random_state=42)
```

### Model Building

```julia
# Using LogisticRegression
@sk_import linear_model: LogisticRegression
```

```julia
model = LogisticRegression(fit_intercept=true)
fit!(model, X_train, y_train)
```

```julia
## Checking The Accuracy of Our Model
accuracy = sum(predict(model, X_test) .== y_test) / length(y_test)
println("Accuracy of model: $accuracy")
```
